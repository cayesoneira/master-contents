```r
#
# 161005 FJC: example of galaxy counts
#
# References
#   http://www.harding.edu/fmccown/r/
#   abline https://stat.ethz.ch/R-manual/R-devel/library/graphics/html/abline.html
#

##########
#
# Initialising all variables
#
rm(list=ls())


#############
#
# Reading simulated numbers of galaxies in 1000 sky patches
#
galaxies<-read.table("galCounts.dat",header=F)
head(galaxies)
class(galaxies)
```

```r
#
# Keeping numbers of galaxies in the vector data: 1000 values
#
data<-galaxies[,1]
class(data)
```

```r
#
# Calculating and keeping statistical values
#
n<-length(data)
nmean<-mean(data)
nsd<-sd(data)
nmin<-min(data)
nmax<-max(data)
#
# Summary and standard deviation
#
summary(data)
nsd
```

```r
#
# Histogram with overplotted gaussian distribution
#
#hist(data, xlab="Number of galaxies",ylab="#",freq=F,ylim=c(0,0.04))
hist(data, xlab="Number of galaxies",ylab="#",breaks=20)
box(which="plot",lty=1)
#
# Gaussians
#
x<-seq(nmin,nmax)
#lines(x,dnorm(x,nmean,nsd))
#lines(x,dnorm(x,nmean,sqrt(nmean)))
curve(10*n*dnorm(x,nmean,nsd),add=T,col="red",lwd=2)
text(c(60),c(130),labels=c("average and sigma"),col="red")
curve(4.5*n*dnorm(x,nmean,sqrt(nmean)),add=T,col="magenta",lty=2,lwd=2)
text(c(110),c(20),labels=c("average and sqrt(average)"),col="magenta")
dev.copy(pdf,file="datCounts.pdf")
dev.off()
```

```r
#
# Figure with values and error bars
#   from
#      http://stackoverflow.com/questions/13032777/scatter-plot-with-error-bars
#
i<-seq(1,n)
edata<-sqrt(data)
plot(data,pch=19,xlim=c(1,n+100),ylim=c(nmin-max(edata),nmax+max(edata)))
arrows(i,data-edata,i,data+edata,length=0.05,angle=90,code=3)
abline(h=nmean,col="red",lty=1,lwd=2)
text(c(1070),c(nmean+5),labels=c("Average"),col="red")
abline(h=nmean+nsd,col="red",lty=2,lwd=2)
text(c(1070),c(nmean+nsd+5),labels=c("+1sigma"),col="red")
abline(h=nmean-nsd,col="red",lty=2,lwd=2)
text(c(1070),c(nmean-nsd-5),labels=c("-1sigma"),col="red")
abline(h=nmean+sqrt(nmean),col="magenta",lty=3,lwd=2)
text(c(1070),c(nmean+sqrt(nmean)+5),labels=c("+sqrt"),col="magenta")
abline(h=nmean-sqrt(nmean),col="magenta",lty=3,lwd=2)
text(c(1070),c(nmean-sqrt(nmean)-5),labels=c("-sqrt"),col="magenta")
dev.copy(pdf,file="datCounts_scatter.pdf")
dev.off()
```
