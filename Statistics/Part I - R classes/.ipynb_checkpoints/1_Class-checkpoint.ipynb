### Exercise 1: Create 3  vectors

- **Vector** containing names of 4 students ("Joe", "Mary", "Tom", "Jen")
- **Vector** containing their heights (1.65,1.70,1.75,1.60)
- **Vector** containing gender of 4 students ("Male","Female","Male","Female")

```r
stu<-c("Joe","Mary","Tom","Jen")
stu
hei<-c(1.65,1.70,1.75,1.60)
hei
gen<-c("Male","Female","Male","Female")
gen
levels(factor(gen))
```

### Exercise 1b: Basic Operations with vectors
    
- Calculate a new vector double the initial vector heights
- Divide all these new heights by 1.5
- Determine the minimun and the maximum of the new heights
- Determine the 1st and 3rd quantile
- Check with the statistical summary

```r
nhei<-2*hei
cat("The double is",nhei,"\n")
nnhei<-nhei/1.5
cat("The division is",nnhei,"\n")
cat("The minimum is",min(nnhei),"\n")
cat("The maximum is",max(nnhei),"\n")
cat("The first quartile is",quantile(nnhei,0.25),"\n")
cat("The third quartile is",quantile(nnhei,0.75),"\n")
cat("The summary is \n",summary(nnhei),"\n")
```

### Exercise 6: Control structures
Write a loop to print whether the numbers from 1 to 10 are even or odd

```r
install.packages('numbers')
library('numbers')

for (i in 1:10) {
    if(mod(i,2)==0) {
        cat(i,"is even\n")
}
    if(mod(i,2)==1) {
        cat(i,"is odd\n")
}
}

# mod(i,2) also ca be written with i%%2
```
