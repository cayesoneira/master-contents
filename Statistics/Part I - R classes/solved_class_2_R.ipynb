## Plotting and Statistics 
 * Read dataset "DatasaurusDozen.tsv" 
 * Select 4 different subsets: "dino", "star", "x_shape", "bullseye" 
 * Calculate mean(x), mean(y), sd(x), sd(y) for every subset and show result in a table
 * Plot Y vs X for every subset

```r
# read data
datasaurus <- read.csv("DatasaurusDozen.tsv", sep="\t")
datasaurus
```

```r
# select subsets
data1 <- datasaurus[datasaurus$dataset == "dino",]
data2 <- datasaurus[datasaurus$dataset == "star",]
data3 <- datasaurus[datasaurus$dataset == "x_shape",]
data4 <- datasaurus[datasaurus$dataset == "bullseye",]
```

```r
# Calculate some basic statistics
meanx <- c(mean(data1$x),mean(data2$x), mean(data3$x), mean(data4$x))
meany <- c(mean(data1$y),mean(data2$y), mean(data3$y), mean(data4$y))
sdx <- c(sd(data1$x),sd(data2$x),sd(data3$x),sd(data4$x))
sdy <- c(sd(data1$y),sd(data2$y),sd(data3$y),sd(data4$y))
stat_table <- data.frame("MeanX"=meanx, "MeanY"=meany, "SDX"=sdx, "SDY"=sdy)
rownames(stat_table) <- c("data1", "data2", "data3", "data4")
stat_table
```

```r
#Plot data
par(mfrow=c(2,2))
par(cex=0.5)
plot(data1$x, data1$y, main="data1", xlab="X", ylab="Y")
plot(data2$x, data2$y, main="data2", xlab="X", ylab="Y")
plot(data3$x, data3$y, main="data3", xlab="X", ylab="Y")
plot(data4$x, data4$y, main="data4", xlab="X", ylab="Y")
```

## Statistical Functions

See [description](http://venus.ifca.unican.es/Rintro/statisticalTreatment.html) at Rintro tutorial.

Make four plots for t-Student (2 by 2) with data between -3 and 3 and representing:

1. Density functions for 1, 5, 20, 30 degrees of freedom (df) + curve for normal distribution (mean=0, sd=1)
2. t Student (df=30) Random numbers (1000) + histogram + t-Student density curve
3. Empirical Cumulative Distribution Function (ECDF) with the random simulated data (df=30)
4. Theoretical Cumulative Distribution Function (CDF)

```r
par(mfrow=c(2,2))

#########################
# a) density function: Student distribution only depends on "df"= degrees of freedom (mean=0, sd=1)
#########################

x<-seq(-3.,3,length=100)
y<-dt(x,df=30)
plot(x,y,type="l",col="cyan",ylab = "Density Function (x)", 
 main="t-Student density function", ylim=c(0.,0.5))
lines(x,dt(x,df=20),col="blue") # === curve(dt(x,df=20),col="blue", add=T)
lines(x,dt(x,df=5),col="green")
lines(x,dt(x,df=1),col="magenta")
points(x,dnorm(x,mean=0,sd=1), cex=0.3, pch=4)# ==== curve(dnorm(x,mean=0,sd=1),add=T)
legend("topleft",c("Distr. t with df=1","Distr. t with df=5", "Distr. t with df=20",
 "Distr. t with df=30", "Distr. Normal(0,1)"),
 col=c("magenta","green","blue","cyan","black"),cex=0.6,
 lty=c(1,1,1,1,NA),pch=c(NA,NA,NA,NA,21),bty="n",y.intersp = 0.7)

#########################
# b) random numbers
#########################

simdata<-rt(1000,df=30)
#hist(simdata, xlab="random t Student", main="Random data (rdmData)") # in frequency
#rug(simdata)
hist(simdata, xlab="random t Student", main="Random data (rdmData)", freq=FALSE) # in density
curve(dt(x,df=30),add=T,col="cyan")

#########################
# c) ECDF
#########################

plot(ecdf(simdata),verticals = TRUE, col.points = "blue", col.hor = "red", 
 col.vert = "bisque",xlab="random t Student", ylab="ECDF(x)", 
 main="ECDF(rdmData)", xlim=c(min(simdata),max(simdata)))

#########################
# d) Th CDF
#########################
x <- simdata
y <- pt(x,df=30)
plot(x,y,ylab="CDF(x)",xlab="random t Student",cex=0.3, 
 main="CDF: pt(rdmData, df=30)",xlim=c(min(simdata),max(simdata)))
```
