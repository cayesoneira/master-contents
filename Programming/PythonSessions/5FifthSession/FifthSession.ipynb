- Mask the galactic plane (from -20 to 20 degrees in latitute)

```python
maps = cmb_map[0]

# Get the number of pixels in the map
npix = hp.nside2npix(hp.get_nside(maps))

# Get the pixel indices
indices = np.arange(npix)

# Get the pixel coordinates (theta, phi) in radians
theta, phi = hp.pix2ang(hp.get_nside(maps), indices)

# Convert theta to latitude in degrees
latitude = np.degrees(np.pi/2 - theta)

# Create a mask that is True for pixels outside the desired latitude range
mask = (latitude < -20) | (latitude > 20)

m_map = hp.ma(maps)
m_map.mask = ~mask

# Plot the masked map
hp.mollview(m_map)
```

- Compute the power-spectra of the map and compare it with the map generated in class (from theoretical cls)

```python
fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(15,5))

plt.axes(ax1)
hp.mollview(cmb_map[0],title='original',cmap='Spectral_r',hold=True)

plt.axes(ax2)
hp.mollview(new_map_cls[0],title=r'from $C_{\ell}$',cmap='Spectral_r',hold=True)
```
