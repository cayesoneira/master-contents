## Q2: *Transform the first CMD (app), i.e., V vs. B-V, into a CMD^ (app). You have to remove the reddening by interstellar extinction, i.e., correct CMD (app). Table 6 of Layden & Sarajedini displays details on the dust extinction at the positions of 54 RRab Lyrae stars. The E(B-V) values range from 0.18 to 0.37 mag, so there is a variable reddening across the field of view with <E(B-V)> = 0.264 $\pm$ 0.005 mag (rms = 0.036 mag). At the XY position of each star, find the nearest RRab star. If the RRab is within 30'' of the star (each pixel has a size of 0.3''), apply the reddening of the RRab to the star; if it is not, omit the star from the dereddened CMD. This approach was used by Layden & Sarajedin*

## To clean variable.txt:

```sos
cp variable.txt 'New_variable.txt'
cat variable.txt | grep -v 'c' | sed -e 's/\t/ /g' | tr -s ' ' > Clean_variables.dat
```

### We read the cleaned data of the variable stars:

```sos
Y = pd.read_csv('Clean_variables.dat', sep=" ")
```

### Create arrays to contain the color intensity and dereddening for each variable star:

```sos
IDvar = np.array(Y["ID"])
Bvar = np.array(Y["B"])
Vvar = np.array(Y["V"])
Ivar = np.array(Y["I"])
ABvar = np.array(Y["A_B"])
AVvar = np.array(Y["A_V"])
AIvar = np.array(Y["A_I"])
```

### Create the loop to remove the reddening in stars close to the variables:

```sos
IDnew=0*ID
Vnew=0*V
Bnew=0*B
Inew=0*I

# We are looping in all the stars measured.
for i in range(len(ID)-1):
    # We start calculating a vector with the position for each star.
    Pos=np.array([xpix[i],ypix[i]])
    
    # We restart the initial distance between stars to a very large number
    dist=10000000
    
    # Now we compare our selected star with every variable one:
    for j in range(len(IDvar)-1):
        
        # First place: if the star is variable we assign to it the value tabulated.
        if IDvar[j]==ID[i]:
            IDnew[i]=ID[i]
            Vnew[i]=V[i]-AVvar[j]
            Bnew[i]=B[i]-ABvar[j]
            Inew[i]=I[i]-AIvar[j]
            break
            
        # If the star is not variable we take the position of the variable star:
        VarPos=np.array([xpix[ IDvar[j]+1 ],ypix[ IDvar[j]+1 ]])
        
        #Calculate the distance of the variable to the original star
        provdist=np.linalg.norm(VarPos-Pos)
        
        #And if it is closer to the star than any other variable star seen before we update the values
        if provdist<100 and provdist<dist:
            dist=provdist
            IDnew[i]=ID[i]
            Vnew[i]=V[i]-AVvar[j]
            Bnew[i]=B[i]-ABvar[j]
            Inew[i]=I[i]-AIvar[j]
            continue
```

## Create a file with the dereddened values:

```sos
der=np.transpose(np.matrix([IDnew,Vnew,Bnew,Inew]))
np.savetxt('dered.txt',der)
```

## 3. To clean the final.txt:

```sos
cat dered.txt | grep -v '000000000000000000' | grep -v 'nan' | tr -s '\t' > Clean_final.dat
```

## We import the final dereddened data:

```sos
Z = pd.read_csv('Clean_final.dat', sep=" ",header=None)
Z.columns = ['ID','V','B','I']

IDf = np.array(Z["ID"])
Vf = np.array(Z["V"])
Bf = np.array(Z["B"])
If = np.array(Z["I"])
```

## New pixel position:

```sos
xf=0*IDf
yf=0*IDf

# Conditions: y[(1 < x) & (x < 5)]

for i in range(len(IDf)-1):
    xf[i]=np.array([xpix[i]])
    yf[i]=np.array([ypix[i]])
    
x = xf
y = yf

plt.style.use('dark_background')
fig, ax = plt.subplots()
ax.plot(x, y,marker='o',linewidth=0,markersize=1)
ax.set_xlabel("X pixel")
ax.set_ylabel("Y pixel")
plt.show()
```

## Plot of V vs. B-V for the dereddened CMD^(app)

```sos
x = Bf-Vf
y = Vf

plt.style.use('dark_background')
fig, ax = plt.subplots()
ax.plot(x, y,marker="o",linewidth=0,markersize=1)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel("B-V")
ax.set_ylabel("V")
ax.set_title("Dereddened CMD")
plt.xlim([-0.5, 1.8])
plt.show()
```

## Plot of V vs. V-I for the dereddened CMD^(app)

```sos
x = Vf-If
y = Vf

plt.style.use('dark_background')
fig, ax = plt.subplots()
ax.plot(x, y,marker="o",linewidth=0,markersize=1)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel("V-I")
ax.set_ylabel("V")
ax.set_title("Dereddened CMD")
plt.xlim([-0.5, 1.8])
plt.show()
```
