```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

X = pd.read_csv('chis.dat', sep=" ",header=None)
X.columns=['reff','chi']

alpha = np.array(X["reff"])
chi = np.array(X["chi"])



plt.style.use('seaborn-v0_8')

fig, ax = plt.subplots()
plt.plot(alpha,chi,linewidth=0,markersize=5,marker='o')
ax.set_xlabel("$R_{eff}$")
ax.set_ylabel("$\chi^{2}$")
ax.set_title("$\chi^{2}$ for different $R_{eff}$")
plt.rcParams["figure.figsize"] = (10,6)
plt.ylim([24,32])

plt.savefig('dves.png')
plt.show()
```
