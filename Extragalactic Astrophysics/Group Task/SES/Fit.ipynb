```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

X = pd.read_csv('chis.dat', sep=" ",header=None)
X.columns=['alpha','chi']

alpha = np.array(X["alpha"])
chi = np.array(X["chi"])

plt.style.use('seaborn-v0_8')

fig, ax = plt.subplots()
plt.plot(alpha,chi,linewidth=0,markersize=5,marker='o')
ax.set_xlabel("alpha")
ax.set_ylabel("$\chi^{2}$")
ax.set_title("$\chi^{2}$ for different alpha")
plt.rcParams["figure.figsize"] = (10,6)

plt.savefig('ses.png')
plt.show()
```

```python
x=np.linspace(0.885,1.055,1000)
y=np.linspace(1,1,1000)

a = 0.902
b = 1.0345

fig, ax = plt.subplots()
plt.plot(alpha,chi,linewidth=0,markersize=5,marker='o')
plt.plot(x,y,linewidth=0.5,markersize=2,marker='.',color="black")
plt.plot([a,a], [1.5, -0.05], '--', lw=1,color="black")
plt.plot([b,b], [1.5, -0.05], '--', lw=1,color="black")
ax.set_xlabel("alpha")
ax.set_ylabel("$\chi^{2}$")
ax.set_title("$\chi^{2}$ for different alpha")
plt.rcParams["figure.figsize"] = (10,6)

plt.ylim([-0.1,1.5])
plt.xlim([0.885,1.055])

plt.savefig('sigma.png')
plt.show()
```
