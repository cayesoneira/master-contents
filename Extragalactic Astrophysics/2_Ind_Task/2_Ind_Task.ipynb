```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sc
import numpy as np
from numpy import linalg
```

```python
X = pd.read_csv('templatespec.dat', sep=" ",header=None,skiprows=9)
X.columns=['Lambda','Flux']

temp_l = np.array(X["Lambda"])
temp_f = np.array(X["Flux"])
```

```python
Y = pd.read_csv('ABGspec.dat', sep=" ",header=None,skiprows=10)
Y.columns=['Lambda','Flux A','Flux B','Flux C']

spec_l = np.array(Y["Lambda"])
spec_f = np.array(Y["Flux C"])
```

```python
spec_t=sc.integrate.simps(spec_f, spec_l, dx=1, axis=-1, even='avg')
temp_t=sc.integrate.simps(temp_f, temp_l, dx=1, axis=-1, even='avg')
```

```python
plt.style.use('seaborn-v0_8')
plt.plot(temp_l,temp_f/temp_t,linewidth=0.5,markersize=0,marker='.')
plt.plot(spec_l,spec_f/spec_t,linewidth=0.5,markersize=0,marker='.',c="green")
plt.grid()
#plt.xlim([5500,9000])
plt.show()
```
