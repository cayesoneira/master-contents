## (Execute 1 in *Cleaning.ipynb*)

```python
X = pd.read_csv('Clean_CMD.dat', sep=" ",header=None)
```

```python
X.columns=['ID','X.Pixel','Y.pixel','V.band.magnitude','Error.in.V.band','B.band.magnitude',
           'Error.in.B.band','I.band.magnitude','Error.in.I.band','DAOPHOT.chi.parameter',
           'DAOPHOT.sharp.sigma.parameter']

ID = np.array(X["ID"])

xpix = np.array(X["X.Pixel"])
ypix = np.array(X["Y.pixel"])

V = np.array(X["V.band.magnitude"])
B = np.array(X["B.band.magnitude"])
I = np.array(X["I.band.magnitude"])
```

## (Execute 2 in *Cleaning.ipynb*)

```python
Y = pd.read_csv('Clean_variables.dat', sep=" ")
```

```python
IDvar = np.array(Y["ID"])
Bvar = np.array(Y["B"])
Vvar = np.array(Y["V"])
Ivar = np.array(Y["I"])
ABvar = np.array(Y["A_B"])
AVvar = np.array(Y["A_V"])
AIvar = np.array(Y["A_I"])

IDnew=0*ID
Vnew=0*V
Bnew=0*B
Inew=0*I

# We are looping in all the stars measured.
for i in range(len(ID)-1):
    # We start calculating a vector with the position for each star.
    Pos=np.array([xpix[i],ypix[i]])
    
    # We restart the initial distance to a very large number
    dist=10000000
    
    # Now we check our selected star with every variable one:
    for j in range(len(IDvar)-1):
        
        # First place: if the star is variable we assign to it the value tabulated.
        if IDvar[j]==ID[i]:
            Vnew[i]=Vvar[j]
            Bnew[i]=Bvar[j]
            Inew[i]=Ivar[j]
            break
            
        # If the star is not variable we take the position of the variable star:
        VarPos=np.array([xpix[ IDvar[j]+1 ],ypix[ IDvar[j]+1 ]])
        
        #Calculate the distance of the variable to the original star
        provdist=np.linalg.norm(VarPos-Pos)
        
        #And if it is closer to the star than any other variable star seen before we update the values
        if provdist<100 and provdist<dist:
            dist=provdist
            IDnew[i]=ID[i]
            Vnew[i]=V[i]-AVvar[j]
            Bnew[i]=B[i]-ABvar[j]
            Inew[i]=I[i]-AIvar[j]
```

```python
NEW=np.transpose(np.matrix([IDnew,Vnew,Bnew,Inew]))
NEW.shape
np.savetxt('final.txt',NEW)
```

We start with the transformation of the absolute magnitudes in apparent (d must be in parsecs):

```python
v0 = lambda x: x + 5*np.log10(d)-5
```
